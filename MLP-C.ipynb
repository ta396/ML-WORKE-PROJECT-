```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
data
```

```python
x=data.data
y=data.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_sc_train = sc.fit_transform(x_train)
x_sc_test = sc.transform(x_test)
```

```python
##### MLP ARC 
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
## input layer
model.add(Dense(units=10,activation="relu"))
```

```python
#hidden layer
model.add(Dense(units=10,activation="relu"))
model.add(Dense(units=10,activation="relu"))
model.add(Dense(units=10,activation="relu"))
```

```python
## OUTPUT LAYER
model.add(Dense(units=1,activation="sigmoid")) ### SOFTMAX
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"]) ### sparse categorical crossentropy
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
pred=model.predict(x_test)
```

```python
pred = (pred > 0.5).astype(int)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print(confusion_matrix(y_test,pred))
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
```
