```python
###### Linear regression 
### lib (sklearn)
```

```python
import pandas as pd 
import sklearn
```

```python
df=pd.read_csv("USA_Housing.csv")
```

```python
df.head()
```

```python
x=df.drop("Price",axis=1)### FEATURES
y=df["Price"]### TARGET 
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import r2_score ,mean_squared_error
```

```python
print(r2_score(pred,y_test)**0.5)
```

```python
print(mean_squared_error(pred,y_test)**0.5)
```

```python
###### POLYNOMIAL REG
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
```

```python
x_poly_train=poly.fit_transform(x_train)
x_poly_test=poly.fit_transform(x_test)
```

```python
model=lr.fit(x_poly_train,y_train)
```

```python
pred=model.predict(x_poly_test)
```

```python
pred
```

```python
print(r2_score(pred,y_test)**0.5)
print(mean_squared_error(pred,y_test)**0.5)
```

```python
#### DECISION TREE REGRESSION 
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(max_depth=15)
```

```python
model=dt.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
print(r2_score(pred,y_test)**0.5)
print(mean_squared_error(pred,y_test)**0.5)
```

```python
###### K-Nearest Neignbor (KNN)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor(n_neighbors=6)
```

```python
model=knn.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
print(r2_score(pred,y_test)**0.5)
print(mean_squared_error(pred,y_test)**0.5)
```

```python
##### SUPPORT VECTOR MACHINES (SVRegressor)
```

```python
from sklearn.svm import SVR
```

```python
sv=SVR()
```

```python
model=sv.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```
